```python
import math
#import numpy as np
from PyQt5 import QtCore, QtGui, QtWidgets

class Ui_Form(object):
    def __init__(self, Form):
        #initialize various variables with their default values
        self.title = "Crossbill Design: design your own single objective light-sheet microscope"
        self.designTLFL1 = 180
        self.M1 = 20
        self.NA1 = 1.00 #numerical aperture
        self.n1 = 1.333 #immersion media refractive index
        self.designTLFL2 = 180 
        self.M2 = 20
        self.NA2 = 0.75
        self.n2 = 1.0
        self.TLFL1 = 200 #tube lens focal length
        self.TLFL2 = 150  
        self.SLFL1 = 100 #scan lens focal length
        self.SLFL2 = 100
        self.designTLFL3 = 180 
        self.M3 = 10
        self.NA3 = 0.45
        self.n3 = 1.0
        self.px = 5.86 #camera pixel size in µm
        self.pxw = 1920 #pixels along width
        self.pxh = 1200 #pixels along height        
        self.TLFL3 = 80
        self.SLFL3 = 100
        self.tilt = 45 #light-sheet tilt angle in degrees
        self.FN1 = 26.5 #Field number of MO1
        self.FN2 = 26.5
        self.FN3 = 26.5
        self.Herschel = False
        self.Nyquist = False
        self.offset = 0 #beam offset in mm
        self.NAx = 0 #effective system NA along x-axis
        self.NAy = 0
        self.M = 10
        self.interM = 1.0 #magnification till intermediate plane
        self.setupUi(Form)

    def setupUi(self, Form):
        Form.setObjectName("Form")
        Form.resize(882, 575)
        self.gridLayout_2 = QtWidgets.QGridLayout(Form)
        self.gridLayout_2.setObjectName("gridLayout_2")
        spacerItem = QtWidgets.QSpacerItem(20, 40, QtWidgets.QSizePolicy.Minimum, QtWidgets.QSizePolicy.Expanding)
        self.gridLayout_2.addItem(spacerItem, 2, 4, 1, 1)
        spacerItem1 = QtWidgets.QSpacerItem(20, 40, QtWidgets.QSizePolicy.Minimum, QtWidgets.QSizePolicy.Expanding)
        self.gridLayout_2.addItem(spacerItem1, 4, 4, 1, 1)
        spacerItem2 = QtWidgets.QSpacerItem(20, 40, QtWidgets.QSizePolicy.Minimum, QtWidgets.QSizePolicy.Expanding)
        self.gridLayout_2.addItem(spacerItem2, 6, 4, 1, 1)
        spacerItem3 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Expanding, QtWidgets.QSizePolicy.Minimum)
        self.gridLayout_2.addItem(spacerItem3, 7, 2, 1, 4)
        spacerItem4 = QtWidgets.QSpacerItem(20, 40, QtWidgets.QSizePolicy.Minimum, QtWidgets.QSizePolicy.Expanding)
        self.gridLayout_2.addItem(spacerItem4, 0, 4, 1, 1)
        self.gridLayout_9 = QtWidgets.QGridLayout()
        self.gridLayout_9.setObjectName("gridLayout_9")
        self.label_22 = QtWidgets.QLabel(Form)
        self.label_22.setObjectName("label_22")
        self.gridLayout_9.addWidget(self.label_22, 1, 1, 1, 1)
        self.lineEdit_pxh = QtWidgets.QLineEdit(Form)
        self.lineEdit_pxh.setObjectName("lineEdit_pxh")
        self.gridLayout_9.addWidget(self.lineEdit_pxh, 4, 2, 1, 1)
        self.lineEdit_pxw = QtWidgets.QLineEdit(Form)
        self.lineEdit_pxw.setObjectName("lineEdit_pxw")
        self.gridLayout_9.addWidget(self.lineEdit_pxw, 4, 1, 1, 1)
        self.lineEdit_px = QtWidgets.QLineEdit(Form)
        self.lineEdit_px.setObjectName("lineEdit_px")
        self.gridLayout_9.addWidget(self.lineEdit_px, 2, 2, 1, 1)
        self.label_20 = QtWidgets.QLabel(Form)
        self.label_20.setObjectName("label_20")
        self.gridLayout_9.addWidget(self.label_20, 0, 1, 1, 2, QtCore.Qt.AlignHCenter)
        self.lineEdit_TL3 = QtWidgets.QLineEdit(Form)
        self.lineEdit_TL3.setObjectName("lineEdit_TL3")
        self.gridLayout_9.addWidget(self.lineEdit_TL3, 2, 1, 1, 1)
        self.label_21 = QtWidgets.QLabel(Form)
        self.label_21.setObjectName("label_21")
        self.gridLayout_9.addWidget(self.label_21, 1, 2, 1, 1)
        self.label_25 = QtWidgets.QLabel(Form)
        self.label_25.setObjectName("label_25")
        self.gridLayout_9.addWidget(self.label_25, 3, 1, 1, 1)
        self.label_26 = QtWidgets.QLabel(Form)
        self.label_26.setObjectName("label_26")
        self.gridLayout_9.addWidget(self.label_26, 3, 2, 1, 1)
        spacerItem5 = QtWidgets.QSpacerItem(20, 40, QtWidgets.QSizePolicy.Minimum, QtWidgets.QSizePolicy.Expanding)
        self.gridLayout_9.addItem(spacerItem5, 0, 0, 5, 1)
        self.gridLayout_2.addLayout(self.gridLayout_9, 4, 3, 1, 1)
        self.gridLayout_5 = QtWidgets.QGridLayout()
        self.gridLayout_5.setObjectName("gridLayout_5")
        self.pushButton_1 = QtWidgets.QPushButton(Form)        
        self.pushButton_1.setObjectName("pushButton_1")
        self.gridLayout_5.addWidget(self.pushButton_1, 0, 0, 1, 1)
        self.plainTextEdit_1 = QtWidgets.QPlainTextEdit(Form)
        self.plainTextEdit_1.setObjectName("plainTextEdit_1")
        self.gridLayout_5.addWidget(self.plainTextEdit_1, 1, 0, 1, 1)
        self.gridLayout_2.addLayout(self.gridLayout_5, 0, 5, 1, 1)
        self.gridLayout_4 = QtWidgets.QGridLayout()
        self.gridLayout_4.setObjectName("gridLayout_4")
        self.comboBox_M2 = QtWidgets.QComboBox(Form)
        self.comboBox_M2.setEditable(True) #make combobox editable
        self.comboBox_M2.setObjectName("comboBox_M2")
        self.comboBox_M2.addItem("")
        self.comboBox_M2.addItem("")
        self.comboBox_M2.addItem("")
        self.comboBox_M2.addItem("")
        self.comboBox_M2.addItem("")
        self.comboBox_M2.addItem("")
        self.comboBox_M2.addItem("")        
        self.gridLayout_4.addWidget(self.comboBox_M2, 2, 2, 1, 1)
        self.label_3 = QtWidgets.QLabel(Form)
        self.label_3.setObjectName("label_3")
        self.gridLayout_4.addWidget(self.label_3, 0, 1, 1, 2, QtCore.Qt.AlignHCenter)
        self.label_10 = QtWidgets.QLabel(Form)
        self.label_10.setObjectName("label_10")
        self.gridLayout_4.addWidget(self.label_10, 3, 1, 1, 1)
        self.lineEdit_NA2 = QtWidgets.QLineEdit(Form)
        self.lineEdit_NA2.setObjectName("lineEdit_NA2")
        self.gridLayout_4.addWidget(self.lineEdit_NA2, 4, 1, 1, 1)
        self.label_9 = QtWidgets.QLabel(Form)
        self.label_9.setObjectName("label_9")
        self.gridLayout_4.addWidget(self.label_9, 1, 2, 1, 1)
        self.comboBox_MO2 = QtWidgets.QComboBox(Form)
        self.comboBox_MO2.setObjectName("comboBox_MO2")
        self.comboBox_MO2.addItem("")
        self.comboBox_MO2.addItem("")
        self.comboBox_MO2.addItem("")
        self.comboBox_MO2.addItem("")
        self.comboBox_MO2.addItem("")
        self.comboBox_MO2.addItem("")
        self.comboBox_MO2.addItem("")
        self.comboBox_MO2.addItem("")
        self.gridLayout_4.addWidget(self.comboBox_MO2, 2, 1, 1, 1)
        self.label_8 = QtWidgets.QLabel(Form)
        self.label_8.setObjectName("label_8")
        self.gridLayout_4.addWidget(self.label_8, 1, 1, 1, 1)
        self.lineEdit_n2 = QtWidgets.QLineEdit(Form)
        self.lineEdit_n2.setObjectName("lineEdit_n2")
        self.gridLayout_4.addWidget(self.lineEdit_n2, 4, 2, 1, 1)
        self.label_11 = QtWidgets.QLabel(Form)
        self.label_11.setObjectName("label_11")
        self.gridLayout_4.addWidget(self.label_11, 3, 2, 1, 1)
        spacerItem6 = QtWidgets.QSpacerItem(20, 40, QtWidgets.QSizePolicy.Minimum, QtWidgets.QSizePolicy.Expanding)
        self.gridLayout_4.addItem(spacerItem6, 0, 0, 5, 1)
        self.gridLayout_2.addLayout(self.gridLayout_4, 0, 3, 1, 1)
        self.gridLayout_13 = QtWidgets.QGridLayout()
        self.gridLayout_13.setObjectName("gridLayout_13")
        self.label = QtWidgets.QLabel(Form)        
        self.label.setObjectName("label")
        self.gridLayout_13.addWidget(self.label, 0, 0, 1, 1, QtCore.Qt.AlignHCenter)
        self.gridLayout_2.addLayout(self.gridLayout_13, 8, 2, 1, 4)
        spacerItem7 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Expanding, QtWidgets.QSizePolicy.Minimum)
        self.gridLayout_2.addItem(spacerItem7, 3, 2, 1, 4)
        spacerItem8 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Expanding, QtWidgets.QSizePolicy.Minimum)
        self.gridLayout_2.addItem(spacerItem8, 5, 2, 1, 4)
        self.gridLayout_11 = QtWidgets.QGridLayout()
        self.gridLayout_11.setObjectName("gridLayout_11")
        self.label_32 = QtWidgets.QLabel(Form)
        self.label_32.setObjectName("label_32")
        self.gridLayout_11.addWidget(self.label_32, 1, 5, 1, 1)
        self.label_28 = QtWidgets.QLabel(Form)
        self.label_28.setObjectName("label_28")
        self.gridLayout_11.addWidget(self.label_28, 1, 1, 1, 1)
        self.label_29 = QtWidgets.QLabel(Form)
        self.label_29.setObjectName("label_29")
        self.gridLayout_11.addWidget(self.label_29, 1, 2, 1, 1)
        self.label_27 = QtWidgets.QLabel(Form)
        self.label_27.setObjectName("label_27")
        self.gridLayout_11.addWidget(self.label_27, 0, 1, 1, 5, QtCore.Qt.AlignHCenter)
        self.label_31 = QtWidgets.QLabel(Form)
        self.label_31.setObjectName("label_31")
        self.gridLayout_11.addWidget(self.label_31, 1, 4, 1, 1)
        self.label_30 = QtWidgets.QLabel(Form)
        self.label_30.setObjectName("label_30")
        self.gridLayout_11.addWidget(self.label_30, 1, 3, 1, 1)
        self.lineEdit_SL3 = QtWidgets.QLineEdit(Form)
        self.lineEdit_SL3.setObjectName("lineEdit_SL3")
        self.gridLayout_11.addWidget(self.lineEdit_SL3, 2, 1, 1, 1)
        self.lineEdit_tilt = QtWidgets.QLineEdit(Form)
        self.lineEdit_tilt.setObjectName("lineEdit_tilt")
        self.gridLayout_11.addWidget(self.lineEdit_tilt, 2, 2, 1, 1)
        self.lineEdit_FN1 = QtWidgets.QLineEdit(Form)
        self.lineEdit_FN1.setObjectName("lineEdit_FN1")
        self.gridLayout_11.addWidget(self.lineEdit_FN1, 2, 3, 1, 1)
        self.lineEdit_FN2 = QtWidgets.QLineEdit(Form)
        self.lineEdit_FN2.setObjectName("lineEdit_FN2")
        self.gridLayout_11.addWidget(self.lineEdit_FN2, 2, 4, 1, 1)
        self.lineEdit_FN3 = QtWidgets.QLineEdit(Form)
        self.lineEdit_FN3.setObjectName("lineEdit_FN3")
        self.gridLayout_11.addWidget(self.lineEdit_FN3, 2, 5, 1, 1)
        self.gridLayout_2.addLayout(self.gridLayout_11, 6, 2, 1, 2)
        spacerItem9 = QtWidgets.QSpacerItem(40, 20, QtWidgets.QSizePolicy.Expanding, QtWidgets.QSizePolicy.Minimum)
        self.gridLayout_2.addItem(spacerItem9, 1, 2, 1, 4)
        self.gridLayout_7 = QtWidgets.QGridLayout()
        self.gridLayout_7.setObjectName("gridLayout_7")
        self.pushButton_2 = QtWidgets.QPushButton(Form)
        self.pushButton_2.setObjectName("pushButton_2")
        self.gridLayout_7.addWidget(self.pushButton_2, 0, 0, 1, 1)
        self.plainTextEdit_2 = QtWidgets.QPlainTextEdit(Form)
        self.plainTextEdit_2.setObjectName("plainTextEdit_2")
        self.gridLayout_7.addWidget(self.plainTextEdit_2, 1, 0, 1, 1)
        self.gridLayout_2.addLayout(self.gridLayout_7, 2, 5, 1, 1)
        self.gridLayout_6 = QtWidgets.QGridLayout()
        self.gridLayout_6.setObjectName("gridLayout_6")
        self.label_12 = QtWidgets.QLabel(Form)
        sizePolicy = QtWidgets.QSizePolicy(QtWidgets.QSizePolicy.Preferred, QtWidgets.QSizePolicy.Minimum)
        sizePolicy.setHorizontalStretch(0)
        sizePolicy.setVerticalStretch(0)
        sizePolicy.setHeightForWidth(self.label_12.sizePolicy().hasHeightForWidth())
        self.label_12.setSizePolicy(sizePolicy)        
        self.label_12.setObjectName("label_12")
        self.gridLayout_6.addWidget(self.label_12, 0, 0, 1, 4, QtCore.Qt.AlignHCenter)
        self.label_14 = QtWidgets.QLabel(Form)
        self.label_14.setObjectName("label_14")
        self.gridLayout_6.addWidget(self.label_14, 1, 2, 1, 1)
        self.label_16 = QtWidgets.QLabel(Form)
        self.label_16.setObjectName("label_16")
        self.gridLayout_6.addWidget(self.label_16, 1, 0, 1, 1)
        self.label_15 = QtWidgets.QLabel(Form)
        self.label_15.setObjectName("label_15")
        self.gridLayout_6.addWidget(self.label_15, 1, 1, 1, 1)
        self.label_13 = QtWidgets.QLabel(Form)
        self.label_13.setObjectName("label_13")
        self.gridLayout_6.addWidget(self.label_13, 1, 3, 1, 1)
        self.lineEdit_TL1 = QtWidgets.QLineEdit(Form)
        self.lineEdit_TL1.setObjectName("lineEdit_TL1")
        self.gridLayout_6.addWidget(self.lineEdit_TL1, 2, 0, 1, 1)
        self.lineEdit_TL2 = QtWidgets.QLineEdit(Form)
        self.lineEdit_TL2.setObjectName("lineEdit_TL2")
        self.gridLayout_6.addWidget(self.lineEdit_TL2, 2, 1, 1, 1)
        self.lineEdit_SL1 = QtWidgets.QLineEdit(Form)
        self.lineEdit_SL1.setObjectName("lineEdit_SL1")
        self.gridLayout_6.addWidget(self.lineEdit_SL1, 2, 2, 1, 1)
        self.lineEdit_SL2 = QtWidgets.QLineEdit(Form)
        self.lineEdit_SL2.setObjectName("lineEdit_SL2")
        self.gridLayout_6.addWidget(self.lineEdit_SL2, 2, 3, 1, 1)
        self.gridLayout_2.addLayout(self.gridLayout_6, 2, 2, 1, 2)
        self.gridLayout_10 = QtWidgets.QGridLayout()
        self.gridLayout_10.setObjectName("gridLayout_10")
        self.pushButton_3 = QtWidgets.QPushButton(Form)
        self.pushButton_3.setObjectName("pushButton_3")
        self.gridLayout_10.addWidget(self.pushButton_3, 0, 0, 1, 1)
        self.plainTextEdit_3 = QtWidgets.QPlainTextEdit(Form)
        self.plainTextEdit_3.setObjectName("plainTextEdit_3")
        self.gridLayout_10.addWidget(self.plainTextEdit_3, 1, 0, 1, 1)
        self.gridLayout_2.addLayout(self.gridLayout_10, 4, 5, 1, 1)
        self.gridLayout_8 = QtWidgets.QGridLayout()
        self.gridLayout_8.setObjectName("gridLayout_8")
        self.comboBox_MO3 = QtWidgets.QComboBox(Form)
        self.comboBox_MO3.setObjectName("comboBox_MO3")
        self.comboBox_MO3.addItem("")
        self.comboBox_MO3.addItem("")
        self.comboBox_MO3.addItem("")
        self.comboBox_MO3.addItem("")
        self.comboBox_MO3.addItem("")
        self.comboBox_MO3.addItem("")
        self.comboBox_MO3.addItem("")
        self.comboBox_MO3.addItem("")
        self.gridLayout_8.addWidget(self.comboBox_MO3, 2, 0, 1, 1)
        self.label_17 = QtWidgets.QLabel(Form)
        self.label_17.setObjectName("label_17")
        self.gridLayout_8.addWidget(self.label_17, 0, 0, 1, 2, QtCore.Qt.AlignHCenter)
        self.label_23 = QtWidgets.QLabel(Form)
        self.label_23.setObjectName("label_23")
        self.gridLayout_8.addWidget(self.label_23, 3, 0, 1, 1)
        self.label_19 = QtWidgets.QLabel(Form)
        self.label_19.setObjectName("label_19")
        self.gridLayout_8.addWidget(self.label_19, 1, 1, 1, 1)
        self.comboBox_M3 = QtWidgets.QComboBox(Form)
        self.comboBox_M3.setEditable(True)
        self.comboBox_M3.setObjectName("comboBox_M3")
        self.comboBox_M3.addItem("")
        self.comboBox_M3.addItem("")
        self.comboBox_M3.addItem("")
        self.comboBox_M3.addItem("")
        self.comboBox_M3.addItem("")
        self.comboBox_M3.addItem("")
        self.comboBox_M3.addItem("")
        self.gridLayout_8.addWidget(self.comboBox_M3, 2, 1, 1, 1)
        self.label_18 = QtWidgets.QLabel(Form)
        self.label_18.setObjectName("label_18")
        self.gridLayout_8.addWidget(self.label_18, 1, 0, 1, 1)
        self.label_24 = QtWidgets.QLabel(Form)
        self.label_24.setObjectName("label_24")
        self.gridLayout_8.addWidget(self.label_24, 3, 1, 1, 1)
        self.lineEdit_NA3 = QtWidgets.QLineEdit(Form)
        self.lineEdit_NA3.setObjectName("lineEdit_NA3")
        self.gridLayout_8.addWidget(self.lineEdit_NA3, 4, 0, 1, 1)
        self.lineEdit_n3 = QtWidgets.QLineEdit(Form)
        self.lineEdit_n3.setObjectName("lineEdit_n3")
        self.gridLayout_8.addWidget(self.lineEdit_n3, 4, 1, 1, 1)
        self.gridLayout_2.addLayout(self.gridLayout_8, 4, 2, 1, 1)
        self.gridLayout_12 = QtWidgets.QGridLayout()
        self.gridLayout_12.setObjectName("gridLayout_12")
        self.pushButton_4 = QtWidgets.QPushButton(Form)
        self.pushButton_4.setObjectName("pushButton_4")
        self.gridLayout_12.addWidget(self.pushButton_4, 0, 0, 1, 1)
        self.plainTextEdit_4 = QtWidgets.QPlainTextEdit(Form)
        self.plainTextEdit_4.setObjectName("plainTextEdit_4")
        self.gridLayout_12.addWidget(self.plainTextEdit_4, 1, 0, 1, 1)
        self.gridLayout_2.addLayout(self.gridLayout_12, 6, 5, 1, 1)
        self.gridLayout_3 = QtWidgets.QGridLayout()
        self.gridLayout_3.setObjectName("gridLayout_3")
        self.label_2 = QtWidgets.QLabel(Form)
        self.label_2.setObjectName("label_2")
        self.gridLayout_3.addWidget(self.label_2, 0, 0, 1, 2, QtCore.Qt.AlignHCenter)
        self.lineEdit_NA1 = QtWidgets.QLineEdit(Form)
        self.lineEdit_NA1.setObjectName("lineEdit_NA1")
        self.gridLayout_3.addWidget(self.lineEdit_NA1, 4, 0, 1, 1)
        self.label_5 = QtWidgets.QLabel(Form)
        self.label_5.setObjectName("label_5")
        self.gridLayout_3.addWidget(self.label_5, 1, 1, 1, 1)
        self.label_4 = QtWidgets.QLabel(Form)
        self.label_4.setObjectName("label_4")
        self.gridLayout_3.addWidget(self.label_4, 1, 0, 1, 1)
        self.label_6 = QtWidgets.QLabel(Form)
        self.label_6.setObjectName("label_6")
        self.gridLayout_3.addWidget(self.label_6, 3, 0, 1, 1)
        self.comboBox_M1 = QtWidgets.QComboBox(Form)
        self.comboBox_M1.setEditable(True)
        self.comboBox_M1.setObjectName("comboBox_M1")
        self.comboBox_M1.addItem("")
        self.comboBox_M1.addItem("")
        self.comboBox_M1.addItem("")
        self.comboBox_M1.addItem("")
        self.comboBox_M1.addItem("")
        self.comboBox_M1.addItem("")
        self.comboBox_M1.addItem("")
        self.gridLayout_3.addWidget(self.comboBox_M1, 2, 1, 1, 1)
        self.comboBox_MO1 = QtWidgets.QComboBox(Form)
        self.comboBox_MO1.setObjectName("comboBox_MO1")
        self.comboBox_MO1.addItem("")
        self.comboBox_MO1.addItem("")
        self.comboBox_MO1.addItem("")
        self.comboBox_MO1.addItem("")
        self.comboBox_MO1.addItem("")
        self.comboBox_MO1.addItem("")
        self.comboBox_MO1.addItem("")
        self.comboBox_MO1.addItem("")
        self.gridLayout_3.addWidget(self.comboBox_MO1, 2, 0, 1, 1)
        self.label_7 = QtWidgets.QLabel(Form)
        self.label_7.setObjectName("label_7")
        self.gridLayout_3.addWidget(self.label_7, 3, 1, 1, 1)
        self.lineEdit_n1 = QtWidgets.QLineEdit(Form)
        self.lineEdit_n1.setObjectName("lineEdit_n1")
        self.gridLayout_3.addWidget(self.lineEdit_n1, 4, 1, 1, 1)
        self.gridLayout_2.addLayout(self.gridLayout_3, 0, 2, 1, 1)

        self.retranslateUi(Form)
        QtCore.QMetaObject.connectSlotsByName(Form)

    def retranslateUi(self, Form):
        _translate = QtCore.QCoreApplication.translate
        #Form.setWindowTitle(_translate("Form", "Crossbill"))
        Form.setWindowTitle(self.title)
        self.label_22.setText(_translate("Form", "TL3 focal length (mm)"))
        self.lineEdit_pxh.setText(_translate("Form", str(self.pxh)))
        self.lineEdit_pxw.setText(_translate("Form", str(self.pxw)))
        self.lineEdit_px.setText(_translate("Form", str(self.px)))
        self.label_20.setText(_translate("Form", "Select TL3 and camera parameters"))
        self.lineEdit_TL3.setText(_translate("Form", str(self.TLFL3)))
        self.label_21.setText(_translate("Form", "Camera pixel size (μm)"))
        self.label_25.setText(_translate("Form", "Pixels along width"))
        self.label_26.setText(_translate("Form", "Pixels along height"))
        self.pushButton_1.setText(_translate("Form", "Click to register MO1, MO2"))
        self.pushButton_1.setStyleSheet("background-color: #607D8B; color: #FFFFFF") #change color!
        self.plainTextEdit_1.setPlainText(_translate("Form", "Enter MO1 and MO2 parameters on the left. Click the button (above) to register the values in this tool."))
        self.comboBox_M2.setItemText(0, _translate("Form", "4x"))
        self.comboBox_M2.setItemText(1, _translate("Form", "10x"))
        self.comboBox_M2.setItemText(2, _translate("Form", "20x"))
        self.comboBox_M2.setItemText(3, _translate("Form", "40x"))
        self.comboBox_M2.setItemText(4, _translate("Form", "50x"))
        self.comboBox_M2.setItemText(5, _translate("Form", "60x"))
        self.comboBox_M2.setItemText(6, _translate("Form", "100x"))
        self.comboBox_M2.setCurrentIndex(2)
        self.label_3.setText(_translate("Form", "Select MO2 parameters"))
        self.label_10.setText(_translate("Form", "NA"))
        self.lineEdit_NA2.setText(_translate("Form", str(self.NA2)))
        self.label_9.setText(_translate("Form", "Magnification"))
        self.comboBox_MO2.setItemText(0, _translate("Form", "Leica"))
        self.comboBox_MO2.setItemText(1, _translate("Form", "Mitutoyo"))
        self.comboBox_MO2.setItemText(2, _translate("Form", "Nikon"))
        self.comboBox_MO2.setItemText(3, _translate("Form", "Olympus"))
        self.comboBox_MO2.setItemText(4, _translate("Form", "Zeiss"))
        self.comboBox_MO2.setItemText(5, _translate("Form", "165 mm"))
        self.comboBox_MO2.setItemText(6, _translate("Form", "180 mm"))
        self.comboBox_MO2.setItemText(7, _translate("Form", "200 mm"))
        self.comboBox_MO2.setCurrentIndex(3)
        self.label_8.setText(_translate("Form", "Make/design tubelens FL"))
        self.lineEdit_n2.setText(_translate("Form", str(self.n2)))
        self.label_11.setText(_translate("Form", "Immersion media R.I."))
        #urlLink="<a href=\"https://kozorovitskiy.org/\">kozorovitskiy.org</a>" 
        doiLink="<a href=\"https://doi.org/10.5281/zenodo.3543786\">10.5281/zenodo.3543786</a>"
        self.label.setText(_translate("Form", "(Author: Manish Kumar, Kozorovitskiy lab, Northwestern University. Last modified: November, 2019). For more details and how to cite info visit "+doiLink))
        self.label.setOpenExternalLinks(True)
        #self.label.setText(_translate("Form", "Written by Manish Kumar, Kozorovitskiy lab, Northwestern University (Last modified: November, 2019)"))
        self.label_32.setText(_translate("Form", "FN3"))
        self.label_28.setText(_translate("Form", "SL3 focal length (mm)"))
        self.label_29.setText(_translate("Form", "Light-sheet tilt (degrees)"))
        self.label_27.setText(_translate("Form", "Select SL3, light-sheet tilt angle, and MO field-numbers (FN)"))
        self.label_31.setText(_translate("Form", "FN2"))
        self.label_30.setText(_translate("Form", "FN1"))
        self.lineEdit_SL3.setText(_translate("Form", str(self.SLFL3)))
        self.lineEdit_tilt.setText(_translate("Form", str(self.tilt)))
        self.lineEdit_FN1.setText(_translate("Form", str(self.FN1)))
        self.lineEdit_FN2.setText(_translate("Form", str(self.FN2)))
        self.lineEdit_FN3.setText(_translate("Form", str(self.FN3)))
        self.pushButton_2.setText(_translate("Form", "Click to register TL, SL"))
        self.pushButton_2.setStyleSheet("background-color: #607D8B; color: #FFFFFF")
        self.plainTextEdit_2.setPlainText(_translate("Form", "Enter tube lens and scan lens focal length values on the left."
                                                     +" Click the button (above) to register the values in this tool."))
        self.label_12.setText(_translate("Form", "Select tube-lens (TL) and scan-lens (SL) focal lengths"))
        self.label_14.setText(_translate("Form", "SL1 focal length (mm)"))
        self.label_16.setText(_translate("Form", "TL1 focal length (mm)"))
        self.label_15.setText(_translate("Form", "TL2 focal length (mm)"))
        self.label_13.setText(_translate("Form", "SL2 focal length (mm)"))
        self.lineEdit_TL1.setText(_translate("Form", str(self.TLFL1)))
        self.lineEdit_TL2.setText(_translate("Form", str(self.TLFL2)))
        self.lineEdit_SL1.setText(_translate("Form", str(self.SLFL1)))
        self.lineEdit_SL2.setText(_translate("Form", str(self.SLFL2)))
        self.pushButton_3.setText(_translate("Form", "Click to register MO3, TL3, Cam"))
        self.pushButton_3.setStyleSheet("background-color: #607D8B; color: #FFFFFF")
        self.plainTextEdit_3.setPlainText(_translate("Form", "Enter MO3, 3rd tube lens focal length and camera parameters on the left."
                                                     +" Click the button (above) to register the values in this tool."))
        self.comboBox_MO3.setItemText(0, _translate("Form", "Leica"))
        self.comboBox_MO3.setItemText(1, _translate("Form", "Mitutoyo"))
        self.comboBox_MO3.setItemText(2, _translate("Form", "Nikon"))
        self.comboBox_MO3.setItemText(3, _translate("Form", "Olympus"))
        self.comboBox_MO3.setItemText(4, _translate("Form", "Zeiss"))
        self.comboBox_MO3.setItemText(5, _translate("Form", "165 mm"))
        self.comboBox_MO3.setItemText(6, _translate("Form", "180 mm"))
        self.comboBox_MO3.setItemText(7, _translate("Form", "200 mm"))
        self.comboBox_MO3.setCurrentIndex(3)
        self.label_17.setText(_translate("Form", "Select MO3 parameters"))
        self.label_23.setText(_translate("Form", "NA"))
        self.label_19.setText(_translate("Form", "Magnification"))
        self.comboBox_M3.setItemText(0, _translate("Form", "4x"))
        self.comboBox_M3.setItemText(1, _translate("Form", "10x"))
        self.comboBox_M3.setItemText(2, _translate("Form", "20x"))
        self.comboBox_M3.setItemText(3, _translate("Form", "40x"))
        self.comboBox_M3.setItemText(4, _translate("Form", "50x"))
        self.comboBox_M3.setItemText(5, _translate("Form", "60x"))
        self.comboBox_M3.setItemText(6, _translate("Form", "100x"))
        self.comboBox_M3.setCurrentIndex(2)
        self.label_18.setText(_translate("Form", "Make/design tubelens FL"))
        self.label_24.setText(_translate("Form", "Immersion media R.I."))
        self.lineEdit_NA3.setText(_translate("Form", str(self.NA3)))
        self.lineEdit_n3.setText(_translate("Form", str(self.n3)))
        self.pushButton_4.setText(_translate("Form", "Click to register SL3, Tilt, FN"))
        self.pushButton_4.setStyleSheet("background-color: #607D8B; color: #FFFFFF")
        self.plainTextEdit_4.setPlainText(_translate("Form", "Enter 3rd scan lens focal length, light-sheet tilt and objective field numbers (FN) on the left."
                                                     +" Click the button (above) to register the values in this tool."
                                                     +"\n(If FN is unknown, leave them to the default value and ignore estimated FOV)"))
        self.label_2.setText(_translate("Form", "Select MO1 parameters"))
        self.lineEdit_NA1.setText(_translate("Form", str(self.NA1)))
        self.label_5.setText(_translate("Form", "Magnifiction"))
        self.label_4.setText(_translate("Form", "Make/design tubelens FL"))
        self.label_6.setText(_translate("Form", "NA"))
        self.comboBox_M1.setItemText(0, _translate("Form", "4x"))
        self.comboBox_M1.setItemText(1, _translate("Form", "10x"))
        self.comboBox_M1.setItemText(2, _translate("Form", "20x"))
        self.comboBox_M1.setItemText(3, _translate("Form", "40x"))
        self.comboBox_M1.setItemText(4, _translate("Form", "50x"))
        self.comboBox_M1.setItemText(5, _translate("Form", "60x"))
        self.comboBox_M1.setItemText(6, _translate("Form", "100x"))
        self.comboBox_M1.setCurrentIndex(2)
        self.comboBox_MO1.setItemText(0, _translate("Form", "Leica"))
        self.comboBox_MO1.setItemText(1, _translate("Form", "Mitutoyo"))
        self.comboBox_MO1.setItemText(2, _translate("Form", "Nikon"))
        self.comboBox_MO1.setItemText(3, _translate("Form", "Olympus"))
        self.comboBox_MO1.setItemText(4, _translate("Form", "Zeiss"))
        self.comboBox_MO1.setItemText(5, _translate("Form", "165 mm"))
        self.comboBox_MO1.setItemText(6, _translate("Form", "180 mm"))
        self.comboBox_MO1.setItemText(7, _translate("Form", "200 mm"))
        self.comboBox_MO1.setCurrentIndex(3)
        self.label_7.setText(_translate("Form", "Immersion media R.I."))
        self.lineEdit_n1.setText(_translate("Form", str(self.n1)))
        
        
    # Define own functions for validating and updating GUI entries
    
    def update_designTLFL(self,combobox_reference):
        # Update the Make/designTLFL of the objective. Takes comboBox reference and returns the designTLFL
        val = combobox_reference.currentText()
        if (val == "Zeiss") | (val == "165 mm"):
            designTLFL = 165    
        elif (val == "Olympus") | (val == "180 mm"):
            designTLFL = 180
        else:
            designTLFL = 200
        return designTLFL
    
    def update_M(self,combobox_reference):
        # Update magnification of the objective. Takes comboBox reference and returns the magnification
        val = combobox_reference.currentText()
        msg = ""
        try:
            M = float(val.split('x')[0])
        except:
            msg = "Invalid magnification entry. Resetting to 20x. "
            M = 20
            combobox_reference.setCurrentIndex(2)
        return (M, msg)
    
    def update_NA_n(self,NA_lineedit_reference, n_lineedit_reference):
        # Update numerical aperture and immersion media refractive index of the objective. Takes corresponding lineEdit references and returns the NA, n
        
        #First set the R.I. value
        val = n_lineedit_reference.text()  
        msg = ""        
        try:
            n = float(val)
            if (n < 1) | (n > 1.8):
                msg = "R.I. is out of range. Resetting to 1.333. "
                n = 1.333
                n_lineedit_reference.setText("1.333")                
        except:
            #print("Invalid refractive index 1 entry. Resetting to default.")
            msg = "Invalid R.I. entry. Resetting to 1.333. "
            n = 1.333
            n_lineedit_reference.setText("1.333")
        
        # Next, set the NA value
        val = NA_lineedit_reference.text()  
        try:
            NA = float(val)
            if (NA <= 0) | (NA > 0.98*n):
                msg = msg + "NA is out of range. Resetting both NA and n! "
                NA = 1.0
                NA_lineedit_reference.setText("1.0")
                n = 1.333
                n_lineedit_reference.setText("1.333")
        except:            
            msg = msg + "Invalid NA entry. Resetting to 1. "
            NA = 1.0
            NA_lineedit_reference.setText("1.0")
        return (NA, n, msg)
        
    def update_FL_pxw_pxh(self,lineedit_reference):
        # make sure the entry is positive integer before passing it on
        # works for focal lengths and camera pixel numbers
        val = lineedit_reference.text()
        msg = ""
        try:
            val = int(val)
            if val <= 0:
                msg = "Negative entry detected. Resetting the value. "
                val = 100
                lineedit_reference.setText("100")           
        except:            
            msg = msg + "Invalid entry. Resetting the value. "
            val = 100
            lineedit_reference.setText("100")
        return (val, msg)
    
    def update_px_FN(self,lineedit_reference):
        # Update camera pixel size. Make sure the entry is positive float before passing it on
        val = lineedit_reference.text()
        msg = ""
        try:
            val = float(val)
            if val <= 0:
                msg = "Negative entry detected. Resetting the value. "
                val = 5.86
                lineedit_reference.setText("5.86")           
        except:            
            msg = msg + "Invalid entry. Resetting the value. "
            val = 5.86
            lineedit_reference.setText("5.86")
        return (val, msg)
    
    def update_tilt(self,lineedit_reference):
        # Update light-sheet tilt. Make sure the entry is positive float between 0 and max. tilt before passing it on
        val = lineedit_reference.text()
        msg = ""
        try:
            val = float(val)
            if val <= 0:
                msg = "Negative entry detected. Resetting the value. "
                val = 45
                lineedit_reference.setText("45")  
            if val >= math.degrees(math.asin(min(self.NA1/self.n1,self.NA2/self.n2))):
                msg = msg + "Tilt angle can not exceed NA1 and NA2 limited value. Setting to the allowed maximum value. "
                val = round(math.degrees(math.asin(min(self.NA1/self.n1,self.NA2/self.n2))),1)
                lineedit_reference.setText(str(val))
        except:            
            msg = msg + "Invalid entry. Resetting the value. "
            val = 45
            lineedit_reference.setText("45")
        return (val, msg)
```

```python
"""
More functions for performing following analysis based on the entered values
1. NAlimiter among MO1 and MO2, max. light-sheet tilt allowed
2. Herschel condition hinter

3. Herschel condition tester

4. Max. system NA calculator
5. Overall system magnification calculator
6. Camera limited FOV calculator
7. Nyquist sampling tester

8. Light-sheet tilt limited system NA calculator
9. Beam offset calculator
10. Some hints for parts selection based on mechanical considerations. Clearance angle and working distance 
"""
def maxTilt(self):
    if self.NA2/self.n2<self.NA1/self.n1:
        msg = "\n- Effective NA is limited by MO2. "
    else:
        msg = "\n- Effective NA is limited by MO1. "
    tilt = round(math.degrees(math.asin(min(self.NA1/self.n1,self.NA2/self.n2))), 1)
    if tilt < float(ui.lineEdit_tilt.text()):
        ui.lineEdit_tilt.setText(str(tilt))
    msg = msg + "\n- Max. light-sheet tilt = "+str(tilt)+"°"
    msg = msg + "\n- Up next, to satisfy Herschel condition for minimal optical aberration, have (TLFL1/TLFL2)*(SLFL2/SLFL1) = " 
    msg = msg + str( round( (self.n1*self.designTLFL1*self.M2)/(self.n2*self.designTLFL2*self.M1), 2))      
    return msg
    
def Herschel(self):
    f1 = (self.TLFL1/self.TLFL2)*(self.SLFL2/self.SLFL1)
    f2 = (self.n1*self.designTLFL1*self.M2)/(self.n2*self.designTLFL2*self.M1)
    if abs(f1-f2)/f2 > 0.05:
        msg = "\n- Herschel condition NOT satisfield (error ~"+ str(round(abs(f1-f2)*100/f2))+"%)"
        msg = msg+"\n- Subsequent analysis for resolution in this tool would be inaccurate due to additional optical aberrations"
    else:
        msg = "\n- Herschel condition satisfield (error ~"+ str(round(abs(f1-f2)*100/f2))+"%)"
        self.Herschel = True
    return msg

def maxNA(self):
    if self.NA3>=self.n2:
        xangle = 2*min(math.asin(self.NA1/self.n1), math.asin(self.NA2/self.n2))
        yangle = 2*min(math.asin(self.NA1/self.n1), math.asin(self.NA2/self.n2))
    else:
        xangle = math.asin(self.NA3/self.n2) + min(math.asin(self.NA3/self.n2),2*min(math.asin(self.NA1/self.n1), math.asin(self.NA2/self.n2))-math.pi/2)
        yangle = 2*min(math.asin(self.NA1/self.n1), math.asin(self.NA2/self.n2), math.asin(self.NA3/self.n2))
    if xangle<0:
        xangle = 0 #negative xangle means no overlap leading to zero NA
    #xangle = min(math.asin(self.NA3/self.n3),math.asin(self.n2/self.n3)) + min(math.asin(self.n2*math.sin(2*math.asin(min(self.NA1/self.n1,self.NA2/self.n2))-math.pi/2)/self.n3), math.asin(self.NA3/self.n3))
    
    self.interM = (self.M1*self.TLFL1/self.designTLFL1)*(self.SLFL2/self.SLFL1)/(self.M2*self.TLFL2/self.designTLFL2)
    self.NAx = self.n2*math.sin(xangle/2)*self.interM #n2*sin(effective acceptance angle/2)*Magnification till intermediate image plane
    self.NAy = self.n2*math.sin(yangle/2)*self.interM
    msg = "\n- Max. system (NAx,NAy) = ("+str(round(self.NAx,2))+","+str(round(self.NAy,2))+")"
    return msg

def M_camFOV_Nyquist(self):
    self.interM = (self.M1*self.TLFL1/self.designTLFL1)*(self.SLFL2/self.SLFL1)/(self.M2*self.TLFL2/self.designTLFL2)
    self.M = round((self.M1*self.TLFL1/self.designTLFL1)*(self.SLFL2/self.SLFL1)/(self.M2*self.TLFL2/self.designTLFL2)*(self.M3*self.TLFL3/self.designTLFL3) ,2)    
    msg = "\n- Overall system magnification = "+str(self.M)
    msg = msg+"\n- Camera limited FOV = ("+str(round(self.pxw*self.px/self.M,1))+" µm,"+str(round(self.pxh*self.px/self.M,1))+" µm)"
    if self.px<0.5*(0.61*.52/min(self.NAx,self.NAy)*self.M):
        msg = msg + "\n- Nyquist sampling satisfied"
        self.Nyquist = True
    else:
        msg = msg + "\n- Nyquist sampling NOT satisfied"
    return msg
        
def realNA(self):
    #calculation assumes that Herschel condition is satisfied 
    diffTheta = (math.asin(min(self.NA1/self.n1,self.NA2/self.n2)))-math.radians(self.tilt) #in radians
    if self.NA3>=self.n2:
        xangle = 2*min(math.asin(self.NA1/self.n1), math.asin(self.NA2/self.n2))-diffTheta
        yangle = 2*min(math.asin(self.NA1/self.n1), math.asin(self.NA2/self.n2))
    else:
        xangle = math.asin(self.NA3/self.n2) + min(math.asin(self.NA3/self.n2), 2*min(math.asin(self.NA1/self.n1), math.asin(self.NA2/self.n2))-math.pi/2-diffTheta)
        yangle = 2*min(math.asin(self.NA1/self.n1), math.asin(self.NA2/self.n2), math.asin(self.NA3/self.n2))
    if xangle<0:
        xangle = 0 #negative xangle means no overlap leading to zero NA    
    
    self.NAx = self.n2*math.sin(xangle/2)*self.interM #n2*sin(effective acceptance angle)*Magnification till intermediate image place
    self.NAy = self.n2*math.sin(yangle/2)*self.interM
    msg = "\n- Actual (NAx,NAy) = ("+str(round(self.NAx,2))+","+str(round(self.NAy,2))+")"
    return msg
    
def offset_maxFOV_tips(self):
    self.offset = round(self.SLFL3*math.tan(math.radians(self.tilt)*self.designTLFL1*self.SLFL1/(self.M1*self.TLFL1*self.SLFL2)),1)
    msg = "\n- Beam offset = "+str(self.offset)+" mm"
    
    msg = msg + "\n- Max. allowed FOV = "+str(round(min(self.FN1/self.M1, self.FN2/(self.M2*self.interM), self.FN3/(self.M3*self.interM))*1000))+" µm"
    msg = msg + "\n- For mechanical compatibility of MO2 and MO3, confirm that sum of their clearance angles is larger than "+str(90-self.tilt)+"°"
    return msg
```

```python
# Define own functions to register button press events
# Register Button1 press   
def btn1_press(self):    
    ui.designTLFL1 = ui.update_designTLFL(ui.comboBox_MO1)
    (ui.M1, msg) = ui.update_M(ui.comboBox_M1) 
    txt = msg
    (ui.NA1, ui.n1, msg) = ui.update_NA_n(ui.lineEdit_NA1,ui.lineEdit_n1)
    txt += msg    
    ui.designTLFL2 = ui.update_designTLFL(ui.comboBox_MO2)
    (ui.M2, msg) = ui.update_M(ui.comboBox_M2)   
    txt += msg
    (ui.NA2, ui.n2, msg) = ui.update_NA_n(ui.lineEdit_NA2,ui.lineEdit_n2)
    txt += msg    
    msg = "Registered values MO1: "+ui.comboBox_MO1.currentText()+", "+str(ui.M1)+"x, "+str(ui.NA1)+" NA, "+str(ui.n1)+" R.I., and "
    msg += "MO2: "+ui.comboBox_MO2.currentText()+", "+str(ui.M2)+"x, "+str(ui.NA2)+" NA, "+str(ui.n2)+" R.I."    
    msg += maxTilt(ui)
    msg += "\n\n (If required, scroll up to see full text!)"
    txt += msg
    ui.plainTextEdit_1.setPlainText(txt)    
    
# Register Button2 press   
def btn2_press(self): 
    btn1_press(ui) #simulate button1 press programmatically
    (ui.TLFL1, msg) = ui.update_FL_pxw_pxh(ui.lineEdit_TL1)
    txt = msg
    (ui.TLFL2, msg) = ui.update_FL_pxw_pxh(ui.lineEdit_TL2)
    txt += msg
    (ui.SLFL1, msg) = ui.update_FL_pxw_pxh(ui.lineEdit_SL1)
    txt += msg
    (ui.SLFL2, msg) = ui.update_FL_pxw_pxh(ui.lineEdit_SL2)
    txt += msg
    msg = "Registered focal length values are \n(TL1, TL2, SL1, SL2) = ("+str(ui.TLFL1)+" mm, "+str(ui.TLFL2)+" mm, "+str(ui.SLFL1)+" mm, "+str(ui.SLFL2)+" mm)"
    msg += Herschel(ui)
    msg += "\n\n (If required, scroll up to see full text!)"
    txt += msg
    ui.plainTextEdit_2.setPlainText(txt) 
    
def btn3_press(self):
    btn1_press(ui)
    btn2_press(ui)
    ui.designTLFL3 = ui.update_designTLFL(ui.comboBox_MO3)
    (ui.M3, msg) = ui.update_M(ui.comboBox_M3) 
    txt = msg
    (ui.NA3, ui.n3, msg) = ui.update_NA_n(ui.lineEdit_NA3,ui.lineEdit_n3)
    txt += msg
    (ui.TLFL3, msg) = ui.update_FL_pxw_pxh(ui.lineEdit_TL3)
    txt += msg    
    (ui.px, msg) = ui.update_px_FN(ui.lineEdit_px)
    txt += msg    
    (ui.pxw, msg) = ui.update_FL_pxw_pxh(ui.lineEdit_pxw)
    txt += msg    
    (ui.pxh, msg) = ui.update_FL_pxw_pxh(ui.lineEdit_pxh)
    txt += msg    
    msg = "Registered values MO3: ("+ui.comboBox_MO3.currentText()+", "+str(ui.M3)+"x, "+str(ui.NA3)+" NA, "+str(ui.n3)+" R.I.), "
    msg += "Focal length TL3 = "+str(ui.TLFL3)+" mm, and Camera: ("+str(ui.px)+" µm, "+str(ui.pxw)+" x "+str(ui.pxh)+")"
    msg += maxNA(ui) 
    if ui.NAx!=0:
        msg += M_camFOV_Nyquist(ui) 
    else:
        msg += "\n Design failure! Please choose better objectives."
    msg += "\n\n (If required, scroll up to see full text!)"
    txt += msg
    ui.plainTextEdit_3.setPlainText(txt) 
    
def btn4_press(self):
    btn1_press(ui)
    btn2_press(ui)
    btn3_press(ui)
    (ui.SLFL3, msg) = ui.update_FL_pxw_pxh(ui.lineEdit_SL3)
    txt = msg
    (ui.tilt, msg) = ui.update_tilt(ui.lineEdit_tilt)
    txt += msg
    (ui.FN1, msg) = ui.update_px_FN(ui.lineEdit_FN1)
    txt += msg
    (ui.FN2, msg) = ui.update_px_FN(ui.lineEdit_FN2)
    txt += msg
    (ui.FN3, msg) = ui.update_px_FN(ui.lineEdit_FN3)
    txt += msg
    msg = "Registered values Focal length SL3 = "+str(ui.SLFL3)+" mm, Tilt angle = "+str(ui.tilt)+"°"
    msg += ", and Field numbers (FN1, FN2, FN3) = (" +str(ui.FN1)+","+str(ui.FN2)+","+str(ui.FN3)+")"
    msg += realNA(ui)
    if ui.NAx!=0:
        msg += offset_maxFOV_tips(ui)
    else:
        msg += "\n Design failure! Please choose better objectives."    
    msg += "\n\n (If required, scroll up to see full text!)"
    txt += msg
    ui.plainTextEdit_4.setPlainText(txt) 
    summary_writer(self) #comment this line if output summary file is not desired
    
def summary_writer(self):
    #write a summary file with all information: MO1: 
    import datetime
    currentDT = datetime.datetime.now()    
    txt = "\n\n######################################################################"
    txt += "\nNew design log "+currentDT.strftime("%b %d, %Y")+"; "+currentDT.strftime("%H:%M:%S")  
    txt += "\nHerschel condition satisfied? :"+str(ui.Herschel)
    txt += "\nNyquist's sampling satisfied? :"+str(ui.Nyquist)+"†"
    txt += "\nMO1: "+ui.comboBox_MO1.currentText()+", "+str(ui.M1)+"x, "+str(ui.NA1)+" NA, "+str(ui.n1)+" R.I., "+str(ui.FN1)+" FN"
    txt += "\nMO2: "+ui.comboBox_MO2.currentText()+", "+str(ui.M2)+"x, "+str(ui.NA2)+" NA, "+str(ui.n2)+" R.I., "+str(ui.FN2)+" FN"
    txt += "\nMO3: "+ui.comboBox_MO3.currentText()+", "+str(ui.M3)+"x, "+str(ui.NA3)+" NA, "+str(ui.n3)+" R.I., "+str(ui.FN3)+" FN"
    txt += "\nTube lens (focal lengths): "+str(ui.TLFL1)+" mm, "+str(ui.TLFL2)+" mm, "+str(ui.TLFL3)+" mm"
    txt += "\nScan lens (focal lengths): "+str(ui.SLFL1)+" mm, "+str(ui.SLFL2)+" mm, "+str(ui.SLFL3)+" mm"
    txt += "\nCamera: "+str(ui.px)+" µm, "+str(ui.pxw)+" x "+str(ui.pxh)        
    txt += "\nOverall system NA (NAx,NAy) = ("+str(round(ui.NAx,2))+","+str(round(ui.NAy,2))+")*"
    txt += "\nOverall system magnification = "+str(ui.M)+"x"
    
    FNlimitedFOV = min(ui.FN1/ui.M1, ui.FN2/(ui.M2*ui.interM), ui.FN3/(ui.M3*ui.interM))*1000
    CAMlimitedFOVwidth = ui.pxw*ui.px/ui.M
    CAMlimitedFOVheight = ui.pxh*ui.px/ui.M   
    txt += "\nOverall system FOV = ("+str(round(min(CAMlimitedFOVwidth,FNlimitedFOV),1))+" µm,"+str(round(min(CAMlimitedFOVheight,FNlimitedFOV),1))+" µm)‡"
    if FNlimitedFOV<max(CAMlimitedFOVwidth,CAMlimitedFOVheight):
        txt += " - Limited by microscope objective's field number"
    else:
        txt += " - Limited by camera (can be improved by smarter choice of camera/MO3/tube-lense)"    
    txt += "\nLight-sheet tilt angle = "+str(ui.tilt)+"° , and corresponding lateral beam-offset = "+str(ui.offset)+" mm"
    txt += "\nFor mechanical compatibility of MO2 and MO3, confirm that sum of their clearance angles is larger than "+str(90-ui.tilt)+"°"
    
    txt += "\n\n†Need not strictly satisfy if Herschel condition is not met, or a large field of view with the given camera is a priority."
    txt += "\n*Accurate only if Herschel condition is satisfied."
    txt += "\n‡Accuracy depends on the validity of the entered F.N. of the objectives."

    
    f = open("CrossbillDesign_summary.txt", "a")
    f.write(txt)
    f.close()
```
